# Convert CSV to sqlite

```python
import pandas as pd
pd.options.display.max_columns = None

df = pd.read_csv("powerplants.csv")
df.head(2)
```

```python
df.dtypes
```

```python
import sqlite3
import os

# Create energy.sqlite3 database
conn = sqlite3.connect('energy.sqlite3')

# Write the data to a sqlite table
df.to_sql('powerplants', conn, if_exists='replace', index=False)

conn.close()
```
